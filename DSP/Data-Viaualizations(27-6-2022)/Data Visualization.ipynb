```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
a=pd.read_csv("delhi_whether.csv")
b=pd.read_csv("kolkata_weather.csv")
fig,ax=plt.subplots()
ax.plot(a.index,a['MLY-PRCP-NORMAL'],marker="*")
ax.plot(b.index,b['MLY-PRCP-NORMAL'],marker="<")
ax.set_xlabel('Time(months)')
ax.set_ylabel('Perciption(Inches)')
ax.set_title('Weather patterns in Delhi and Kolkata')
plt.show()
```

# Dual axis line chart

```python
plt.style.use('default')
c=pd.read_csv("climate_change.csv")
fig,ax=plt.subplots()
ax.plot(c.index,c['co2'], label='CO2 (ppm)')
# Plot the relative temperature in red
ax.plot(c.index, c['relative_temp'], label='Relative Temp (C)')
ax.set_xlabel('Time (years)')
ax.set_title('CO2 and Temperature over the Years')
ax.legend()
plt.show()
```

# Plotting two variables-twin axes

```python
fig,ax=plt.subplots()
ax.plot(c.index,c['co2'], label='CO2 (ppm)',c="#377eb8")
ax.set_xlabel('Time (years)',fontsize=17,labelpad=15)
ax.set_ylabel('CO2(ppm)',c="#377eb8",fontsize=17,labelpad=15)
ax.tick_params(axis='y',labelsize=15,colors="#377eb8")
ax2=ax.twinx()
ax2.plot(c.index, c['relative_temp'], label='Relative Temp (C)',c="#e41a1c")
ax2.set_ylabel('Relative temperature',c="#e41a1c",fontsize=17,labelpad=15)
ax2.tick_params(axis='y',labelsize=15,colors="#e41a1c")
ax2.set_title('CO2 and Temperature over the Years',fontsize=19,pad=15,c='darkgreen')
plt.show()
```

# Dual axis line chart with annotations

```python
#with plt.style.context('seaborn'):
fig, ax = plt.subplots()
# Plot the relative temperature data
ax.plot(c.index, c['relative_temp'], c='#377eb8')
ax.set_xlabel('Time (years)', fontsize=17, labelpad=15)
ax.set_ylabel('Relative Temperature (Celsius)', fontsize=17, labelpad=15)
ax.tick_params(axis='both', labelsize=15)
#annotate the date at which temperature exceeded 1 degree
ax.annotate('>1-Degree',fontsize=14,
              xy=(pd.to_datetime('2015-10-06'),1),
              xytext=(pd.to_datetime('2008-10-06'),0.1),
              arrowprops={'arrowstyle':'simple','color':'#e41a1c'})
plt.show()
```

```python
starbucks_survey = pd.read_csv("starbucks_customer_survey.csv")
age_percentages = starbucks_survey['2. Your Age'].value_counts(normalize=True) * 100

colors = ['#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']

fig, ax = plt.subplots()

ax.set_title('Age Groups of Starbucks Customers', fontsize=14, pad=15)

# Plot pie chart
ax.pie(x=age_percentages, labels=age_percentages.index, colors=colors, autopct='%.1f%%')

# bbox_to_anchor keyword gives a great degree of control for manual legend placement
ax.legend(loc='upper right', bbox_to_anchor=(1.4, 1), fontsize=10)

plt.show()
```

```python
countries_of_the_world=pd.read_csv("countries_of_the_world.csv")
fig, ax = plt.subplots()
# Plot scatter chart
ax.scatter(countries_of_the_world['Life expectancy'], countries_of_the_world['GDP per capita'])
ax.set_title('Life Expectancy vs GDP per capita of Countries 2015', fontsize=10, pad=15)
ax.set_xlabel('Life Expectancy', fontsize=8, labelpad=10)
ax.set_ylabel('GDP per capita ($)', fontsize=8, labelpad=10)
ax.tick_params(axis='both', labelsize=5)
plt.show()
```

```python
fig, ax = plt.subplots()

i = 0
colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33']
groups = countries_of_the_world.groupby('Continent')
for name, group in groups:
    ax.scatter(group['Life expectancy'], group['GDP per capita'], label=name, c=colors[i]) # Plot scatter chart for each continent
    i +=1
ax.set_title('Life Expectancy vs GDP per capita of Countries 2015', fontsize=20, pad=15)

ax.set_xlabel('Life Expectancy', fontsize=16, labelpad=10)
ax.set_ylabel('GDP per capita ($)', fontsize=16, labelpad=10)
ax.tick_params(axis='both', labelsize=12)

ax.legend()

plt.show()
```

```python
#bar(c,v)
summer_olympics=pd.read_csv("summer_olympics.csv")
summer_olympics['Sport'].unique()
gymnastics = summer_olympics[summer_olympics['Sport'] == 'Gymnastics']
gymnastics_2012 = gymnastics[gymnastics['Year'] == 2012]
gymnastics_2012_disciplines = gymnastics_2012['Discipline'].unique()
no_of_athletes_per_discipline = gymnastics_2012['Discipline'].value_counts()
fig, ax = plt.subplots()

# Plot a bar chart
ax.bar(gymnastics_2012_disciplines, no_of_athletes_per_discipline)

ax.set_xticklabels(['Artistic', 'Rhythmic', 'Trampoline'], fontsize=15)

ax.set_ylabel('Number of Athletes', fontsize=17, labelpad=15)
ax.tick_params(axis='y', labelsize=15)

ax.set_title('Gymnastics Disciplines (Summer Olympics 2012)', fontsize=18, pad=17)

plt.show()
```

```python
commonwealth_medals=pd.read_csv("commonwealth_games_medals.csv",index_col=0)
fig, ax = plt.subplots()

# Plot a horizontal bar chart of gold medals as a function of country and change the color of the bars
ax.barh(commonwealth_medals.index, commonwealth_medals['Gold'], color='#4daf4a')

ax.set_title('All Time Medals Won by Countries at Commonwealth Games', fontsize=20, pad=25)
ax.set_xlabel('Number of medals', fontsize=18, labelpad=10)
ax.tick_params(axis='both', labelsize=15)

plt.show()
```

```python
five_countries_medals = commonwealth_medals[:5]
fig, ax = plt.subplots()

# Add bars for 'Gold' with the label 'Gold'
ax.bar(five_countries_medals.index, five_countries_medals['Gold'], label='Gold', color='#377eb8')

# Stack bars for 'Silver' on top with label 'Silver'
ax.bar(five_countries_medals.index, five_countries_medals['Silver'], bottom=five_countries_medals['Gold'], label='Silver', color='#ff7f00')

# Stack bars for 'Bronze' on top of that with label 'Bronze'
ax.bar(five_countries_medals.index, five_countries_medals['Bronze'], bottom=five_countries_medals['Gold'] + five_countries_medals['Silver'], label='Bronze', color='#4daf4a')

ax.set_title('Medals Won By Countries', pad=15, fontsize=17)

ax.set_xlabel('Country', labelpad=15, fontsize=13)
ax.tick_params(axis='x', rotation=45)
ax.set_ylabel('Number of Medals', labelpad=15, fontsize=13)

# Display the legend
ax.legend()

plt.show()
```

```python
#hist(v,bins=)
cancer_m=pd.read_csv("cancer_m.csv")
fig, ax = plt.subplots()

# Default number of bins is 10, use the 'bins' argument to change it
ax.hist(cancer_m['Area (mean)'],bins=10,ec='black')

ax.set_xlabel('Area (mean)', fontsize=15, labelpad=15)
ax.set_ylabel('Number of observations', fontsize=15, labelpad=15)
ax.tick_params(axis='both', labelsize=13)
ax.set_title('Area Distibution of Malignant Cancer Tumors', fontsize=16, pad=15)

plt.show()
```

```python
delhi_weather=pd.read_csv("delhi_whether.csv")
kolkata_weather=pd.read_csv("kolkata_weather.csv")
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Adjust the size of the figure
fig.set_size_inches([10, 6])

# Addressing the top left Axes as index 0, 0, plot month and Delhi precipitation
ax[0, 0].plot(delhi_weather.index, delhi_weather['MLY-PRCP-NORMAL'])
ax[0, 0].set_title('Delhi Precipitation')

# In the top right (index 0,1), plot month and Delhi temperatures
ax[0, 1].plot(delhi_weather.index, delhi_weather['MLY-TAVG-NORMAL'])
ax[0, 1].set_title('Delhi Temperature')




# In the bottom left (1, 0) plot month and Kolkata precipitations
ax[1, 0].plot(kolkata_weather.index, kolkata_weather['MLY-PRCP-NORMAL'])
ax[1, 0].set_title('Kolkata Precipitation')

# In the bottom right (1, 1) plot month and Kolkata temperatures
ax[1, 1].plot(kolkata_weather.index, kolkata_weather['MLY-TAVG-NORMAL'])
ax[1, 1].set_title('Kolkata Temperature')

#Set the title for the entire subplot
fig.suptitle('Weather in Delhi and Kolkata', fontsize=15, fontweight ='bold')

# Automatically adjusts the padding between and around subplots
fig.tight_layout()

# Adjust the positioning
fig.subplots_adjust(top=0.85, wspace=0.15, hspace=0.5)

plt.show()
```
