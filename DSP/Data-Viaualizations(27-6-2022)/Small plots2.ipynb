```python
import matplotlib.pyplot as plt
import numpy as np
#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,9,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("sjfksnjfls")
#plot2
x=np.array([0,1,2,3])
y=np.array([30,80,90,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.pie(y)
plt.title("hdkjsds")
plt.suptitle("ajndlsjadol")
plt.show()
```
