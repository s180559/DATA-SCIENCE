```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import svm
x=np.array([[1,3],[1,2],[1,1.5],[1.5,2],[2,3],[2.5,1.5],[2,1],[3,1],[3,2],[3.5,1],[3.5,3]])
y=[0]*6+[1]*5
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=50,alpha=0.9)
```

```python
svm=svm.SVC(kernel='linear').fit(x,y)
```

```python
X,Y=np.mgrid[0:4:200j,0:4:200j]
```

```python
Z=svm.decision_function(np.c_[X.ravel(),Y.ravel()])
```

```python
Z=Z.reshape(X.shape)
```

```python
plt.contourf(X,Y,Z>0,alpha=0.6)
plt.contour(X,Y,Z,colors=['k'],linestyles=['-'],levels=[0])
plt.scatter(x[:,0],x[:,1],c=y,s=50,alpha=0.9)
```

```python
plt.contourf(X,Y,Z>0,alpha=0.9)
plt.contour(X,Y,Z,colors=['k','k','k'],linestyles=['--','-','--'],levels=[-1,0,1])
plt.scatter(svm.support_vectors_[:,0],svm.support_vectors_[:,1],s=120,facecolors='None')
plt.scatter(x[:,0],x[:,1],c=y,s=50,alpha=0.9)
```
